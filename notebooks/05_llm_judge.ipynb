# LLM-Judge Evaluation Notebook
This notebook contains the Large Language Model (LLM) evaluations of the search engine results (TF-IDF, Embedding, Hybrid).  
Each query is evaluated on a 0–3 relevance scale, where:
- **0** = not relevant  
- **1** = weakly relevant  
- **2** = moderately relevant  
- **3** = highly relevant  

The evaluations were performed using an LLM (ChatGPT).


## Query 1 — "latest technology trends"

**Interpretation:**
- Embedding search returned the most relevant articles.
- Hybrid search was moderately relevant.
- TF-IDF returned the least relevant results.
